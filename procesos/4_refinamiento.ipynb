```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
```

```python
# funcin para describir los tpicos
# carrera, palabras de cada tpico y docs representativos
#def generar_titulo_topico(carrera, topicos):
def generar_titulo_topico(carrera, topicos, palabras, docs):
  template= """
  Eres un experto en anlisis temtico de investigaciones acadmicas.
  A partir de los datos que provienen del modelado temtico de trabajos finales de la carrera {carrera},
  debes generar un nombre del tema general o principal del tpico y una explicacin del mismo.
  Tu respuesta debe:
  - estar escrita en idioma espaol
  - relacionada explcitamente con temas tpicos de la carrera {carrera}
  - no incluir ttulos o subttulos para estructurar la respuesta.

  Adems,
  el tema debe ser:
  - descriptivo pero conciso (mximo 7 palabras)
  - NO incluir el smbolo dos puntos ni la palabra 'tpico'
  - Evitar comenzar con alguna de estas palabras: 'Anlisis', 'Estudio','Investigacin'.
  - Usar las palabras clave del tpico: {palabras_clave}.
  - No incluir smbolos especiales

  y la explicacin del tema debe ser:
  - descriptiva pero concisa (mximo 50 palabras)
  - Puedes usar la informacin de los documentos representativos: {doc_repres} solo para la redaccin de la explicacin.

  Proporciona tu respuesta (nombre de tema y explicacin) como texto separado por un salto de lnea .
  Por ejemplo:
  <tu ttulo sugerido aqu>
  <tu explicacin o descripcin aqu>
  """

  prompt = ChatPromptTemplate.from_template(template)

  #model = OllamaLLM(model="llama3.2")
  model = OllamaLLM(model="deepseek-r1")

  chain = prompt | model

  # Prepare a list to store dictionaries of results
  results_data = []

  for k in range(len(topicos)):
    #if topicos['topic'][k]==-1:
    if topicos[k]==-1:
      title=''
      desc=''
      full_response_string=''
    else:
      #full_response_string = chain.invoke({"carrera": topicos['carrera'][k], "palabras_clave":topicos['representation'][k], "doc_repres": topicos['representative_docs'][k]})
      full_response_string = chain.invoke({"carrera": carrera, "palabras_clave":palabras[k], "doc_repres": docs[k]})
      print(str(k)+" ##" +full_response_string)
      print()
      # post = full_response_string.find("TITLE:")
      # posd = full_response_string.find("DESCRIPC:")

      # title = full_response_string[post+8:posd].strip()
      # desc = full_response_string[posd+12:].replace('\n',' ').strip()

      lista=full_response_string.split("\n")
      title=lista[0].strip()
      desc=lista[1].strip()

      if (len(title)==0) or (len(desc)==0):
        print("ERROR: "+full_response_string)

    results_data.append({"titulo": title, "descripcion": desc, "respuesta":full_response_string})

  return pd.DataFrame(results_data)
```
